### Import Module

```python
import matplotlib.pyplot as plt # module utama ploting data
import numpy as np
```

### Membuat Plotting Sederhana

Membuat sebuah figure yang memiliki sebuah axes; melakukan plotting data pada axes

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

fig, ax = plt.subplots() # membuat sebuah figure dan sebuah axes (default)
ax.plot(x, y) # melakukan plotting data pada axes
```

Alternatif lain, kita bisa langsung memanfaatkan method plot() pada pyplot untuk melakukan plotting

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
```

Catatan :
* Figure adalah container besar
* Axes area gambar
* Figure dapat terdiri dari beberapa axes

Cara untuk membuat figure dan axes

```python
fig = plt.figure() # hanya membuat satu figure tanpa axes, 
# tidak bisa melakukan visualisasi data
```

```python
fig, ax = plt.subplots() # sebuah figure dengan sebuah axes
```

```python
fig, axs = plt.subplots(2, 3) # sebuah figure dengan 2x3 grid axes
# dengan 6 buah object axes
```

Dua cara dalam menggunakan Matplotlib
* OO Style (Object Oriented)
* pyplot Style

### Object Oriented Style untuk Plotting

secara explisit membuat figures beserta axes, dan memanggil methods dari keduanya.

```python
x = np.linspace(0, 2, 100) # total jumlah data poin 100
x
```

```python
fig, ax = plt.subplots() # membuat sebuah figure dan sebuah axes

# plotting tiga varian data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label') # menyertakan x-label pada axes
ax.set_ylabel('y label') # menyertakan y-label pada axes
ax.set_title("Simple Plot") # menyertakan title pada axes
ax.legend() # menyertakan legend atau garis
```

### pyplot Style untuk Plotting

Mengandalkan pyplot untuk membuat dan mengelola figures dan axes, serta menggunakan fungsi pada pyplot untuk melakukan plotting

```python
x = np.linspace(0, 2, 100)
x
```

```python
# plotting tiga varian data pada axes
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label') # menyertakan x-label pada axes
plt.ylabel('y label') # menyertakan y-label pada axes
plt.title("Simple Plot") # menyertakan title pada axes
plt.legend() # menyertakan legend atau garis
```
